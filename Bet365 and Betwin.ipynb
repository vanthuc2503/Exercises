<h1>Data Preprocessing</h1>

```python
#import library:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load data
data = pd.read_csv('11.FootballDataEurope.csv')
data.head()
```

```python
#Set option display all data
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#Sort data by country name, league name, season, stage and time
data.sort_values(by=['country_name', 'league_name', 'season', 'stage', 'date'], inplace=True)
data.head()
```

```python
#Display data info
data.info()
```

```python
#Check null values
data.isnull().sum()
```

```python
#Check duplicate values
data.duplicated().sum()
```

```python
#Drop null value
data = data.dropna()
```

```python
#Check info after delete null value
data.info()
```

```python
#set date column to format date-month-year
data['date'] = pd.to_datetime(data['date']).dt.strftime('%d-%m-%Y')
data.head()
```

```python
#Add weekday column, extract date column and delete date column
data['weekday'] = pd.to_datetime(data['date'], dayfirst=True).dt.day_name()
data[['day', 'month', 'year']] = data['date'].str.split('-', expand=True)
data.drop('date', axis=1, inplace=True)
data.head()
```

<h3>Calculate the correct prediction percentage for bet365 and betwin</h3>

```python
# Number of correct predictions for home team win
bet365_home_win_predictions = ((data['B365H'] < data['B365A']) & (data['home_team_goal'] > data['away_team_goal'])).sum()
betwin_home_win_predictions = ((data['BSH'] < data['BSA']) & (data['home_team_goal'] > data['away_team_goal'])).sum()
```

```python
# Number of correct prediction for away team win
bet365_away_win_predictions = ((data['B365A'] < data['B365H']) & (data['away_team_goal'] > data['home_team_goal'])).sum()
betwin_away_win_predictions = ((data['BSA'] < data['BSH']) & (data['away_team_goal'] > data['home_team_goal'])).sum()
```

```python
# Number of predictions for draw
bet365_draw_predictions = ((data['B365D'] < data['B365H']) & (data['B365D'] < data['B365A']) & (data['home_team_goal'] == data['away_team_goal'])).sum()
betwin_draw_predictions = ((data['BSD'] < data['BSH']) & (data['BSD'] < data['BSA']) & (data['home_team_goal'] == data['away_team_goal'])).sum()
```

```python
correct_prediction_b365 = bet365_home_win_predictions + bet365_away_win_predictions + bet365_draw_predictions
correct_proportion_b365 = correct_prediction_b365/len(data)
print(f"Bet365-proportion of correct predictions: {correct_proportion_b365}")
```

```python
correct_prediction_betwin = betwin_home_win_predictions + betwin_away_win_predictions + betwin_draw_predictions
correct_proportion_betwin = correct_prediction_betwin/len(data)
print(f"Betwin-proportion of correct predictions: {correct_proportion_betwin}")
```

```python
# Data for pie charts
bet365_labels = ['Correct predictions', 'Incorrect predictions']
betwin_labels = ['Correct predictions', 'Incorrect predictions']

bet365_sizes = [correct_proportion_b365, 1 - correct_proportion_b365]
betwin_sizes = [correct_proportion_betwin, 1 - correct_proportion_betwin]

plt.figure(figsize=(12, 6))

# Bet365 Pie Chart
plt.subplot(1, 2, 1)
plt.pie(bet365_sizes, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange'])
plt.title('Bet365 proportion of correct predictions')
plt.legend(bet365_labels, loc="best")

# Betwin Pie Chart
plt.subplot(1, 2, 2)
plt.pie(betwin_sizes, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange'])
plt.title('Betwin proportion of correct predictions')
plt.legend(betwin_labels, loc="best")

plt.show()
```

```python
target_goal_correct = (data['target'] == data['diff_goals']).sum()
percentage_correct_target_goal = target_goal_correct / len(data)
print(f"Percentage of correct target goal: {percentage_correct_target_goal}")
```

```python
target_goal_labels = ['Correct predictions', 'Incorrect predictions']

target_goal_sizes = [percentage_correct_target_goal, 1 - percentage_correct_target_goal]

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.pie(target_goal_sizes, autopct='%1.1f%%', startangle=90, colors=['green', 'red'])
plt.title('Proportion of correct target goal predictions')
plt.legend(target_goal_labels, loc="best")

plt.show()
```
